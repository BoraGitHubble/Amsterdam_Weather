```python
# This code retrieves and displays the current weather and hourly forecast information for Amsterdam for free of charge.

# You will get Current Weather:
#    Time: The timestamp of the current weather data.
#    Temperature: The current temperature in degrees Celsius.
#    Windspeed: The current windspeed in meters per second.import requests

# You will get Hourly Forecast and for each hourly forecast data point, the following information is displayed:
#    Time: The timestamp of the forecasted data.
#    Temperature: The forecasted temperature for that hour in degrees Celsius.
#    Relative Humidity: The forecasted relative humidity for that hour as a percentage.
#    Windspeed: The forecasted windspeed for that hour in meters per second.

def weather(latitude, longitude):
    base_url = "https://api.open-meteo.com/v1/forecast"

    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current_weather": "true",
        "hourly": "temperature_2m,relativehumidity_2m,windspeed_10m"
    }

    response = requests.get(base_url, params=params)
    weather_data = response.json()

    if response.status_code == 200 and "current_weather" in weather_data and "hourly" in weather_data:
        current_weather = weather_data["current_weather"]
        hourly_forecast = weather_data["hourly"]

        print("Current Weather:")
        print("Time:", current_weather["time"])
        print("Temperature:", current_weather["temperature"], "°C")
        print("Windspeed:", current_weather["windspeed"], "m/s")
        print("----------------------------------")
        print("Hourly Forecast:")

        for i in range(len(hourly_forecast["time"])):
            time = hourly_forecast["time"][i]
            temperature = hourly_forecast["temperature_2m"][i]
            humidity = hourly_forecast["relativehumidity_2m"][i]
            windspeed = hourly_forecast["windspeed_10m"][i]

            print("Time:", time)
            print("Temperature:", temperature, "°C")
            print("Relative Humidity:", humidity, "%")
            print("Windspeed:", windspeed, "m/s")
            print("----------------------------------")
    else:
        print("Unable to retrieve weather information.")

latitude = 52.377956
longitude = 4.897070
weather(latitude, longitude)
```
